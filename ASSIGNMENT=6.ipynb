```python
#TASK=2
def fibonacci(lim):
  first_num=0
  sec_num=1
  print(first_num,sec_num,end=" ")
  while first_num+sec_num<=lim:
    third_num=first_num+sec_num
    print(third_num,end=" ")
    first_num=sec_num
    sec_num=third_num
fibonacci(8)
```

```python
#TASK=10
def make_square(var,var1):
  out={}
  start=var
  end=var1
  for i in range(start,end+1):
    out.update({i:i**2})
  return out
make_square(1,3)
```

```python
#TASK=13
def function_name(operator,firstOperand,secondOperand):
  if operator=="+":
    res_cal=firstOperand+secondOperand
  elif operator=="-":
    res_cal=firstOperand-secondOperand
  elif operator=="*":
    res_cal=firstOperand*secondOperand
  elif operator=="/":
    res_cal=firstOperand/secondOperand
  return res_cal
operator=input()
firstOperand=float(input())
secondOperand=float(input())
print(function_name(operator,firstOperand,secondOperand))
```

```python
#TASK=14
def function_name(sentence,position):
  primary=""
  remaining_part=""
  if position==0:
    return sentence
  elif position!=0:
    for a in range(0,position):
      primary=primary+sentence[a]

    for i in range(position,len(sentence)):
      if i%position!=0:
        primary=primary+sentence[i]
      if i%position!=0:
        remaining_part=remaining_part+sentence[i]
  storage=primary+remaining_part
  return storage


sentence=input()
position=int(input())
print(function_name(sentence,position))
```
