```python
#TASK=16
s = input()
numbers = [int(x) for x in s.split(", ")]
max = numbers[0]
for i in range(len(numbers)):
  if numbers[i] < max:
    second_max = numbers[i]
    break
for num in numbers:
  if max < num:
    max, second_max = num, max
  elif second_max < num:
    second_max = num
print("My list:", numbers)
print("Second largest number in the list is", second_max, "which was found at index", str(numbers.index(second_max))+'.')
```

```python
#TASK=21
#Method 1
#Removing values from the same input list
s = input()
li = [int(x.strip()) for x in s.split(",")]
print("Given numbers in list:", li)
for i in range(len(li)-1, -1, -1):
  if li[i] in li[:i]:
    li.pop(i)
print("List without any dupliacte values:", li)
```

```python
#TASK=21
#Method 2
#Creating new list and making sure no duplicate values are added
s = input()
li = [int(x.strip()) for x in s.split(",")]
l2 =[]
for i in li:
    if i not in l2:
        l2.append(i)
print("Given numbers in list:", li)
print("List without any dupliacte values:", l2)
```
