```python
#TASK=1
def bubble_sort(numbers):
  for i in range(0, len(numbers)-1):
    for j in range(0,len(numbers)-i-1):
      if numbers[j]> numbers[j+1]:
        temp = numbers[j]
        numbers[j] = numbers[j+1]
        numbers[j+1] = temp

  return numbers

bubble_sorted = bubble_sort([10, 1, 20, 3, 6, 2, 5, 11, 15, 2, 12, 14, 17, 18, 29])
print(bubble_sorted)
```

```python
#TASK=2
def my_list(numbers):
  for index in range(len(numbers)-1):
    min = numbers[index]
    min_index = index
    #find minimum and swap with numbers[index]
    for j in range(index+1, len(numbers)):
         if numbers[j] < min:
               min = numbers[j]
               min_index = j

     #swapping numbers[min_index] with numbers[index]
    temp =  numbers[min_index]
    numbers[min_index] =  numbers[index]
    numbers[index] = temp
  return numbers

numbers = [10, 1, 20, 3, 6, 2, 5, 11, 15, 2, 12, 14, 17, 18, 29]

sorted_numbers = my_list(numbers)
print(sorted_numbers)
```

```python
#TASK=3
def selection_desc(numbers):
  for index in range(len(numbers)-1):
    min = numbers[index]
    min_index = index
    #find maximum and swap with numbers[index]
    for j in range(index+1, len(numbers)):
         if numbers[j] > min:
               min = numbers[j]
               min_index = j

     #swapping numbers[min_index] with numbers[index]
    temp =  numbers[min_index]
    numbers[min_index] =  numbers[index]
    numbers[index] = temp
  return numbers

numbers = [10, 1, 20, 3, 6, 2, 5, 11, 15, 2, 12, 14, 17, 18, 29]
sorted_numbers = selection_desc(numbers)
print(sorted_numbers)
```
